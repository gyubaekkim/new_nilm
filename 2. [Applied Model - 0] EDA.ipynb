```python
from nilmtk import DataSet
from nilmtk.utils import print_dict

from matplotlib import rcParams
import matplotlib.pyplot as plt

%matplotlib inline
rcParams['figure.figsize'] = (12, 8)
```

```python
enertalk = DataSet('../enertalk/enertalk-converted/enertalk.h5')
print_dict(enertalk.metadata)
# must specify the period since enertalk dataset is too big
#enertalk.set_window(start='2016-11-02 07:00:00', end='2016-11-02 08:00:00')
enertalk.set_window(start='2017-01-01 00:00:00', end='2017-02-01 00:00:00')
```

```python
# enertalk.set_window(start='2017-01-01 00:00:00', end='2017-01-15 00:00:00')
# 1, 2, 9, 10, 13, 14, 18, 22

# enertalk.set_window(start='2016-10-01 00:00:00', end='2016-10-15 00:00:00')
# 3, 5, 6

# enertalk.set_window(start='2016-12-01 00:00:00', end='2016-12-15 00:00:00')
# 9

# enertalk.set_window(start='2017-04-01 00:00:00', end='2017-04-15 00:00:00')
# 12, 21

# no rice cooker
# 4, 7, 8, 11, 15, 16, 17, 19, 20

i_list = [1, 2, 3, 5, 6, 9, 10, 12, 13, 14, 18, 21, 22]
```

```python
enertalk.set_window(start='2017-01-03 00:00:00', end='2017-01-05 12:00:00')
elec = enertalk.buildings[1].elec

main = elec.all_meters()[0] # The aggregated meter that provides the input
appliance = elec.submeters()['rice cooker']

main.plot()
appliance.plot()
plt.show()
```

```python
enertalk.set_window(start='2016-04-01 00:00:00', end='2017-04-30 00:00:00')

for i in i_list:
    
    try:
        print(i)
        elec = enertalk.buildings[i].elec

        main = elec.all_meters()[0] # The aggregated meter that provides the input
        appliance = elec.submeters()['rice cooker']

        main.plot()
        appliance.plot()
        plt.show()
        
    except:
        continue
```
